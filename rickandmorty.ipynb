## Using Next

```python
page_num = 40
has_next = True
data = []
while has_next:
    print("ingesting page {}".format(page_num))
    response = requests.get('https://rickandmortyapi.com/api/character?page=' + str(page_num)).json()
    data.extend(response['results'])
    has_next = response['info']['next']
    page_num += 1
```

## Using Pages

```python
page_num = 40
has_next = True
data = []
while has_next:
    print("ingesting page {}".format(page_num))
    response = requests.get('https://rickandmortyapi.com/api/character?page=' + str(page_num)).json()
    data.extend(response['results'])
    has_next = page_num < response['info']['pages']
    page_num += 1 
```

## Using Count

```python
page_num = 1
has_next = True
data = []
while has_next:
    #print("ingesting page {}".format(page_num))
    response = requests.get('https://rickandmortyapi.com/api/character?page=' + str(page_num)).json()
    data.extend(response['results'])
    has_next = len(data) < response['info']['count']
    #has_next = response['info']['next']
    page_num += 1
```

## Without using next, pages, and count

```python
page_num = 40
data = []

while True:
    response = requests.get('https://rickandmortyapi.com/api/character?page=' + str(page_num)).json()

    try: 
        print(len(response['results'][0]))
    except:
        print('End loop')
        break
    data.extend(response['results'])
    page_num += 1
```
